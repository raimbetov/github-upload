```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.random.rand(10)
y = np.random.rand(10)
```

```python
df_x = pd.DataFrame(x, columns = ["Values X"])
df_y = pd.DataFrame(y, columns = ["Values Y"])
```

```python
df_xy = pd.concat([df_x, df_y], axis = 1)
df_xy["Categorical"] = ["A", "B", "A", "B", "A", "A", "B", "B", "B", "A"]
df_xy
```

```python
from scipy import stats
```

```python
sns.set_style("ticks")
sns.barplot(data = df_xy, x = "Categorical", y = "Values X", ci = "sd", color = "slateblue", errwidth = 1.25, capsize = 0.05)
sns.despine()
sns.set_context("poster")
```

```python
stats.ttest_ind(df_xy["Values X"], df_xy["Values Y"]) 
```

```python
sns.jointplot(data = df_xy, x = "Values X", y = "Values Y", color = "slateblue")
```

```python
sns.boxplot(data = df_xy)
sns.despine()
```

```python
x4 = np.random.randint(500, size = 100)
y4 = np.random.randint(500, size = 100)
z4 = np.random.randint(500, size = 100)
```

```python
x4d = pd.DataFrame(x4, columns = ["Values X"])
y4d = pd.DataFrame(y4, columns = ["Values Y"])
z4d = pd.DataFrame(z4*0.75, columns = ["Values Z"])
```

```python
df_xyz = pd.concat([x4d, y4d, z4d], axis = 1)
```

```python
sns.pointplot(data = df_xyz)
```

```python
sns.jointplot(data = df_xyz, x = "Values X", y = "Values Z", kind = "reg").annotate(stats.pearsonr)
```

```python
stats.ttest_ind(df_xyz["Values X"], df_xyz["Values Y"])
```

```python
stats.ttest_ind(df_xyz["Values X"], df_xyz["Values Z"])
```

```python
stats.ttest_ind(df_xyz["Values Y"], df_xyz["Values Z"])
```
